```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from selenium.webdriver.common.action_chains import ActionChains
```

```python
PATH = 'C:\edgeDriver\msedgedriver.exe'
driver = webdriver.Edge(PATH)

driver.get('https://www.cnnindonesia.com/search/?query=penyakit')

berita = []
links = []
dicts = {}

try:
    a = 1
    for paginations in range(2, 7):
        for j in range(1, 11):
            title_xpath = '//*[@id="content"]/div/div[2]/section/div/div[1]/article[' + str(j) + ']/a/span[2]/h2'
            linkxpath = '//*[@id="content"]/div/div[2]/section/div/div[1]/article[' + str(j) + ']/a'
            article = WebDriverWait(driver, 2).until(
                EC.presence_of_element_located((By.XPATH, title_xpath))
            )
            link = WebDriverWait(driver, 2).until(
                EC.presence_of_element_located((By.XPATH, link_xpath))
            )
            href = link.get_attribute('href')
            
            print("Judul Berita", a, ":", article.text, ":", href)
            
            berita.append(article.text)
            links.append(href)
            dicts[article.text] = href
            
            a+=1
            
        pagination = driver.find_element(By.XPATH, 
                                         '//*[@id="content"]/div/div[2]/section/div/div[2]/div/a[' + str(paginations) + ']')
        pagination.click()
        time.sleep(1) 


except Exception as e:
    print('error: ', e)
    
    
driver.quit()
```

```python
df = pd.DataFrame({'Judul Berita': berita, 'Link': links})
df.index = range(1, len(df) + 1)
```
